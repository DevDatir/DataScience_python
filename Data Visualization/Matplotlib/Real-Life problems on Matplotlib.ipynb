### Gas Prices plot


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv("gas_prices.csv")
```

```python
gas
```

```python
print(gas)
```

```python
# plt.plot(gas.Year,gas.Australia, "b.-" ,label="Australia")
# plt.plot(gas.Year,gas.USA, "r.-" ,label="USA")
# plt.plot(gas.Year,gas.UK, "y.-" ,label="UK")
# plt.plot(gas.Year,gas['South Korea'], "g.-" ,label="South Korea")

plt.figure(figsize=(9,4),dpi=300)

for country in gas:
    if country!="Year":
        plt.plot(gas.Year,gas[country], label=country, marker=".",markersize=5)

plt.title("Gas prices Over Time in USD/Gallon",fontdict={"fontsize":15,"fontweight":"bold"})
plt.xlabel("Year",fontdict={"fontsize":15})
plt.ylabel("USD/Gallon",fontdict={"fontsize":14})

plt.xticks(gas.Year[::3].tolist()+[2011])


plt.legend()

plt.savefig("gas_prices_plot.png")

plt.show()
```

```python
plt.hist(fifa.Overall,bins=xt1,color="#abcdef")
xt1 = np.arange(70,100,5)
plt.title("Fifa overall Skill Histogram",fontdict={"fontsize":17,"fontweight":"bold"})
plt.xlabel("Overall",fontdict={"fontsize":14})
plt.ylabel("No. of players",fontdict={"fontsize":14})
plt.xticks(xt1)
#plt.legend()
plt.show()
```

```python
left = fifa.loc[fifa["Preferred Foot"]=="Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"]=="Right"].count()[0]
```

```python
plt.figure(figsize=(7,7),dpi=100)
colors=["#F9EFDB","#910A67"]

labels=["Left","Right"]
plt.title("Foot Preference of FIFA Playes",fontdict={"fontsize":17,"fontweight":"bold"})

plt.pie([left,right], labels = labels, colors=colors,autopct="%.2f %%")


plt.show()
```

```python
fifa.Weight= [int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight <150].count()[0]
medium = fifa.loc[(fifa.Weight >150) & (fifa.Weight<190)].count()[0]
heavy = fifa.loc[fifa.Weight >=190].count()[0]

weights=[light ,medium ,heavy]
labels = ["Under 150","150 to 250","more than 220"]
plt.pie(weights, labels = labels)

plt.show()
```

```python
plt.style.use("default")
plt.figure(figsize=(5,6))
plt.title("FIFA Club Comparison", fontdict={"fontsize":15})
plt.ylabel("Overall", fontdict={"fontsize":15})

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
milan = fifa.loc[fifa.Club == "Milan"]["Overall"]

label = ["FC Barcelona","Real Madrid","AC Milan"]

boxes = plt.boxplot([barcelona,madrid,milan], labels = label, patch_artist=True,medianprops={'linewidth' : 1.5})

for box in boxes["boxes"]:
    box.set(color= "#3652AD",linewidth=1.5)
    
    box.set(facecolor="#E1F0DA")

plt.savefig("fifa_club_comparison.png", dpi = 120)
plt.show()
```
