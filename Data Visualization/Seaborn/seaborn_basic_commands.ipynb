### Seaborn Tutorial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
crash_cs = sns.load_dataset("car_crashes")
```

```python
crash_cs
```

### Distribution Plot

```python
sns.distplot(crash_cs["not_distracted"],kde=False,bins=25)
```

### Joint Plot

```python
sns.jointplot(x="speeding",y="alcohol",data=crash_cs,kind="reg")
```

```python
sns.jointplot(x="speeding",y="alcohol",data=crash_cs,kind="kde")
```

```python
sns.jointplot(x="speeding",y="alcohol",data=crash_cs,kind="hex")
```

## KDE plot

```python
sns.kdeplot(crash_cs["alcohol"])
```

## Pair plots

```python
sns.pairplot(crash_cs)
```

## Categorical Plots

### Bar Plot

```python
plt.figure(figsize=(4,4))
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator=np.median)
```

### Box Plot

```python
sns.boxplot(x="day",y="total_bill",data=tips_df,hue="sex")

plt.legend(loc=0)
```

### Violin Plot

```python
sns.violinplot(x="day",y="total_bill",data=tips_df,hue="sex",split=True)

plt.figure(figsize=(4,4))
```

### Strip PLot

```python
plt.figure(figsize=(6,4))
sns.stripplot(x="day",y="total_bill",data=tips_df,jitter=True,hue="sex",dodge=True)

plt.legend(loc=0)
```
