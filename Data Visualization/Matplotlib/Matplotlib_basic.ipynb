```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1,3,5,7]
y = [2,4,6,8]

plt.figure(figsize=(5,4), dpi=100)

plt.plot(x,y, label = "5x",color="blue",linewidth=2, marker="*",markersize=5,linestyle="--",markeredgecolor = "red")

x2 = np.arange(0,6.5,0.5)

plt.plot(x2[:6],x2[:6]**2,label = "x^2",color = "red",marker="*",markeredgecolor="blue",markersize=5,linewidth=2)

plt.plot(x2[5:],x2[5:]**2, label = "x^2--",color = "red", marker="*", markersize=5,markeredgecolor="blue",linewidth=2,linestyle="--")

plt.title("First Graph", fontdict={"fontname":"Comic Sans MS","fontsize" : 20})

plt.xlabel("X-axis", fontdict={"fontname" : "Comic Sans MS", "fontsize" : 15})
plt.ylabel("Y-axis", fontdict={"fontname" : "Comic Sans MS", "fontsize" : 15})

xt1=np.arange(0,10,1)
yt1=np.arange(0,10,1)
plt.xticks(xt1)
plt.yticks(yt1)

plt.legend()

plt.savefig("mygraph.png", dpi=300)

plt.show()
```

```python
labels = ["A","B","C"]
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=140)

bars = plt.bar(labels,values)

patterns = ["/","O","*"]

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```
